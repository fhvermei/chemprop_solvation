```python
import pandas as pd
from chemprop_solvation.solvation_estimator import load_DirectML_Gsolv_estimator, \
    load_DirectML_Hsolv_estimator, load_SoluteML_estimator
```

```python
# Load all ML models. This step needs to be done only once.
SoluteML_estimator = load_SoluteML_estimator()
Hsolv_estimator = load_DirectML_Hsolv_estimator()
Gsolv_estimator = load_DirectML_Gsolv_estimator()
```

# Example 1. Abraham solute parameters estimation using SoluteML

```python
# Make predictions

solute_smiles = [['CCCCCCCC'],  # solute SMILES 1
          ['Cc1ccccc1']] # solute SMILES 2

avg_pre, epi_unc, valid_indices = SoluteML_estimator(solute_smiles)
```

```python
# Print the results and export the results as an csv file

solute_param_name_list = ['E', 'S', 'A', 'B', 'L']

result_dict = {'smiles': []}
for solute_param_name in solute_param_name_list:
    result_dict[f'{solute_param_name} pred.'] = []
    result_dict[f'{solute_param_name} variance'] = []
    
for i in range(len(solute_smiles)):
    print(f"For {solute_smiles[i][0]}:")
    result_dict['smiles'].append(solute_smiles[i][0])
    for j in range(len(solute_param_name_list)):
        print(f'\t{solute_param_name_list[j]}: {avg_pre[i][j]},  Variance on prediction: {epi_unc[i][j]}')
        result_dict[f'{solute_param_name_list[j]} pred.'].append(avg_pre[i][j])
        result_dict[f'{solute_param_name_list[j]} variance'].append(epi_unc[i][j])

# save as a csv file
df_result = pd.DataFrame(result_dict)
df_result.to_csv('SoluteML_sample_result.csv', index=False)
df_result
```

# Example 2. Solvation free energy (dGsolv) and solvation enthalpy (dHsolv) at 298 K prediction using DirectML

```python
# Make predictions

solv_solute_smiles = [['c1ccccc1', 'Cn1cccc1'], # solvent1, solute1
                      ['C[S](C)=O', 'CCCC']] # solvent2, solute2

Gsolv_avg_pre, Gsolv_epi_unc, Gsolv_valid_indices = Gsolv_estimator(solv_solute_smiles)
Hsolv_avg_pre, Hsolv_epi_unc, Hsolv_valid_indices = Hsolv_estimator(solv_solute_smiles)
```

```python
# Print the results and export the results as an csv file

result_dict = {'solvent smiles': [], 
               'solute smiles': [], 
              'dGsolv pred (kcal/mol)': [],
              'dGsolv variance (kcal/mol)': [],
              'dHsolv pred (kcal/mol)': [],
              'dHsolv variance (kcal/mol)': []}
    
for i in range(len(solv_solute_smiles)):
    solvent_smiles = solv_solute_smiles[i][0]
    solute_smiles = solv_solute_smiles[i][1]
    print(f"For solvent {solvent_smiles} and solute {solute_smiles}:")
    result_dict['solvent smiles'].append(solvent_smiles)
    result_dict['solute smiles'].append(solute_smiles)
    print(f'\tdGsolv (kcal/mol): {Gsolv_avg_pre[i]},  Variance on prediction (kcal/mol): {Gsolv_epi_unc[i]}')
    result_dict['dGsolv pred (kcal/mol)'].append(Gsolv_avg_pre[i])
    result_dict['dGsolv variance (kcal/mol)'].append(Gsolv_epi_unc[i])
    print(f'\tdHsolv (kcal/mol): {Hsolv_avg_pre[i]},  Variance on prediction (kcal/mol): {Hsolv_epi_unc[i]}')
    result_dict['dHsolv pred (kcal/mol)'].append(Hsolv_avg_pre[i])
    result_dict['dHsolv variance (kcal/mol)'].append(Hsolv_epi_unc[i])
    
# save as a csv file
df_result = pd.DataFrame(result_dict)
df_result.to_csv('DirectML_sample_result.csv', index=False)
df_result
```
